# Selenium Web Scraping

### Scrapes fruit information from https://fruitsandveggies.org/ and saves it in JSON format

```python
# Imports
from selenium import webdriver
import os
```

```python
#launch url
url = "https://fruitsandveggies.org/fruits-and-veggies/"

# create a new Chrome session
chrome_path = r'/usr/local/bin/chromedriver' #path from 'which chromedriver'
driver = webdriver.Chrome(executable_path=chrome_path)
driver.get(url)
driver.maximize_window()
```

```python
fruit_lst = ["Apple",
"Avocado",
"Banana",
"Bell Peppers",
"Blackberries",
"Blueberries",
"Cherries (Sweet)",
"Coconut",
"Cucumber",
"Grapes",
"Honeydew Melon",
"Kiwifruit",
"Lemon",
"Lime",
"Lychee",
"Mango",
"Orange",
"Papaya",
"Peach",
"Pear",
"Pineapple",
"Plum",
"Pomegranate",
"Raspberries",
"Strawberries",
"Tomato",
"Watermelon"]
```

```python
fruit_dict = {}

for fruit in fruit_lst:
    # Keep trying each link until it works
    while True:    
        try:
            fruit_link = driver.find_element_by_link_text(fruit)
            fruit_link.click()
            div = None
            try:
                div = driver.find_element_by_class_name('l-grid-item-medium-12')
            except:
                div = driver.find_element_by_class_name('l-grid-item-medium-9')
            fruit_dict[fruit] = div.text
        except:
            continue
        break
    
    driver.back()
```
